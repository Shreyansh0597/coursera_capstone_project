## Analysing and Clustering Neighborhoods in Mérida, Yucatán, México

#### Choosing the best Neighborhood to visit or live

##### Introduction

Mérida is the economic, cultural and service capital of the Mexican southeast. It is a cosmopolitan destination with easy access, connected by air, sea and land with the world thanks to a privileged geographical location. It is a place that enjoys true originality, an expressive richness with universal value and scope. In 2019, the city was designated a Creative Gastronomic City by UNESCO.

In 2020, the Organization for Economic Cooperation and Development (OECD) recognized Mérida as one of the best cities in the country to invest. That same year, CEO World magazine classified Mérida as the second safest city in the American continent only behind Quebec City.

The city of Mérida has been growing and developing significantly and has established itself as a tourist destination worldwide. Due to this, many people from different states of México choose to visit or move and start a new life in this beautiful city. I have decided to perform the following analysis related to the neighborhoods and places of this city in order to provide a guide for those travelers who decide to visit this city and in the same way for those people who have plans to move and start a new life in this beautiful place.
